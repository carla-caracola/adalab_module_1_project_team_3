```python
import random

logo = '''  _   _   _   _   _   _     _   _   _   _   _     _     _   _   _   _   _   _  
 / \ / \ / \ / \ / \ / \   / \ / \ / \ / \ / \   / \   / \ / \ / \ / \ / \ / \ 
( p | i | e | d | r | a ) ( p | a | p | e | l ) ( o ) ( t | i | j | e | r | a )
 \_/ \_/ \_/ \_/ \_/ \_/   \_/ \_/ \_/ \_/ \_/   \_/   \_/ \_/ \_/ \_/ \_/ \_/ '''


# Número de puntos objetivo
puntos_objetivo = 3
puntos_jugador = 0 
puntos_pc = 0
ronda = 1

# Opciones válidas
opciones = ["piedra", "papel", "tijera"]

# Respuestas cuando el jugador gana
jugador_gana_ronda = ["BIEN HECHO!","LA RONDA ES TUYA!","ESTÁS DE SUERTE! HAS GANADO ESTA RONDA"]
jugador_pierde_ronda = ["OOPS...HAS PERDIDO", "TU OPONENTE HA GANADO ESTA PERO, PERO AÚN TIENES CHANCES!", "HAS PERDIDO ESTA RONDA, ÁNIMO QUE EL JUEGO CONTINÚA..."]

print (logo)
print( "\n¡BIENVENIDE A PIEDRA, PAPEL & TIJERA! \n\nEstas son las reglas del juego: \n    1) Dos jugadores eligen una de las tres opciones: Piedra, Papel o Tijera. \n    2) Las reglas para determinar al ganador son: \n       --> Piedra vence a Tijera. \n       --> Tijera vence a Papel. \n       --> Papel vence a Piedra. \n    3) El jugador que gana una ronda obtiene un punto. \n    4) El juego continúa hasta que un jugador haya acumulado 3 puntos, convirtiéndose en el ganador.\n\n¡EMPIEZA EL JUEGO!👇 ")
# Elección del jugador
while puntos_jugador < puntos_objetivo and puntos_pc < puntos_objetivo:

     print(f"\n-------------- RONDA #{ronda}---------------")
     print("\nElige piedra🪨, papel🧻 o tijera✂️...\n")
     eleccion_jugador = input("Elige piedra, papel o tijera: ").lower()

     if eleccion_jugador in opciones:
          ronda += 1
          print("Has elegido:", eleccion_jugador)
     else:
          print("Opción no válida, prueba otra vez.")
          continue
          
     #Eleccion PC
     eleccion_pc = random.choice (opciones)
     print("Tu oponente ha elegido:", eleccion_pc)

     # Decidir ganador
     
     #empate
     if eleccion_jugador == eleccion_pc:
          print("Es un empate!🙃")
          print(f"Tus puntos: {puntos_jugador} \nPuntos oponente: {puntos_pc}")
     #jugador gana
     elif eleccion_jugador == "piedra" and eleccion_pc == "tijera" or eleccion_jugador == "tijera" and eleccion_pc == "papel" or eleccion_jugador == "papel" and eleccion_pc == "piedra":
          print(random.choice(jugador_gana_ronda)) 
          puntos_jugador += 1
     
          print(f"Tus puntos: {puntos_jugador} \nPuntos oponente: {puntos_pc}")
     #jugador pierde
     else: 
          
          print(random.choice(jugador_pierde_ronda))
          puntos_pc += 1
          print(f"Tus puntos: {puntos_jugador} \nPuntos oponente: {puntos_pc}")

else:
     if puntos_jugador == puntos_objetivo:
          print("\n\nFELICIDADES! HAS GANADO EL JUEGO 🏆\n")
     else:
          print("\n\nHAS PERDIDO 😭 MALA SUERTE...INTÉNTALO DE NUEVO!\n")

```
